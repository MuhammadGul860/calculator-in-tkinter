```python
import tkinter
from tkinter import *
```

```python
root=Tk()
root.title("simple calculator")
root.geometry("400x600+100+100")
root.configure(bg="dodger blue")
root.minsize(200,300)
root.maxsize(500,700)


entry_box=Entry(root,width=80,borderwidth=2,bg="#adbaf1",cursor="hand2",font=("ariel",30,"bold"),justify=RIGHT)
entry_box.pack(padx=10,pady=10)

def number_click(number):
    current = entry_box.get()
    entry_box.config(fg="black")
    entry_box.delete(0, END)
    entry_box.insert(0, str(current) + str(number))
    

# Function to clear the entry box
def clear_click():
    entry_box.delete(0, "end")
    
# Function to handle multiplication
def multiply():
    current = entry_box.get()
    entry_box.delete(0, "end")
    entry_box.insert(0, current + "*")

def add():
    current = entry_box.get()
    entry_box.delete(0, "end")
    entry_box.insert(0, current + "+")


def subtract():
    current = entry_box.get()
    entry_box.delete(0, "end")
    entry_box.insert(0, current + "-")


def divide():
    current = entry_box.get()
    entry_box.delete(0, "end")
    entry_box.insert(0, current + "/")





def evaluate():
    try:
        result = eval(entry_box.get())  # Use eval to calculate the result
        entry_box.delete(0, "end")
        entry_box.insert(0, result)
        entry_box.config(fg="black")
    except Exception as e:
        entry_box.delete(0, "end")
        entry_box.insert(0, "Error")
        entry_box.config(fg="red")
        
        

   
# b1=Button(root,text="click",padx=20,pady=10,border=10,borderwidth=5,cursor="hand2")
# b1.pack(pady=5)

b2=Button(root,text="*",padx=20,pady=23,border=5,cursor="hand2",command=multiply)
b2.place(x=280,y=170)
b2=Button(root,text="%",padx=19,pady=23,border=5,cursor="hand2")
b2.place(x=215,y=170)
b2=Button(root,text="/",padx=20,pady=23,border=5,cursor="hand2",command=divide)
b2.place(x=148,y=170)
b2=Button(root,text="C",padx=20,pady=23,bg="lime",border=5,cursor="hand2", command = clear_click)
b2.place(x=80,y=170)

#b2=Button(root,text="",padx=15,pady=20)
#b2.place(x=310,y=455)
b2=Button(root,text="=",padx=18,pady=61,bg="light salmon",border=5,cursor="hand2",command=evaluate)
b2.place(x=280,y=410)
b2=Button(root,text="+",padx=17,pady=23,border=5,cursor="hand2",command=add)
b2.place(x=280,y=326)
b2=Button(root,text="-",padx=20,pady=23,border=4,cursor="hand2",command=subtract)
b2.place(x=280,y=248)

b2=Button(root,text=".",padx=21,pady=24,border=5,cursor="hand2",command=add)
b2.place(x=215,y=485)
b2=Button(root,text="3",padx=20,pady=23,border=5,cursor="hand2", command=lambda: number_click(3))
b2.place(x=215,y=407)
b2=Button(root,text="6",padx=20,pady=23,border=5,cursor="hand2" ,command=lambda: number_click(6))
b2.place(x=215,y=326)
b2=Button(root,text="9",padx=20,pady=23,border=5,cursor="hand2", command=lambda: number_click(9))
b2.place(x=215,y=248)

b2=Button(root,text="0",padx=21,pady=24,border=5,cursor="hand2" ,command=lambda: number_click(0))
b2.place(x=148,y=485)
b2=Button(root,text="2",padx=21,pady=23,border=5,cursor="hand2",command=lambda: number_click(2))
b2.place(x=148,y=407)
b2=Button(root,text="5",padx=20,pady=23,border=5,cursor="hand2", command=lambda: number_click(5))
b2.place(x=148,y=326)
b2=Button(root,text="8",padx=20,pady=23,border=5,cursor="hand2", command=lambda: number_click(8))
b2.place(x=148,y=248)

b2=Button(root,text="00",padx=19,pady=24,border=5,cursor="hand2", command=lambda: number_click("0"+"0"))
b2.place(x=80,y=485)
b2=Button(root,text="1",padx=21,pady=23,border=5,cursor="hand2", command=lambda: number_click(1))
b2.place(x=80,y=407)
b2=Button(root,text="4",padx=21,pady=24,border=5,cursor="hand2", command=lambda: number_click(4))
b2.place(x=80,y=326)
b2=Button(root,text="7",padx=21,pady=23,border=5,cursor="hand2", command=lambda: number_click(7))
b2.place(x=80,y=248)

root.mainloop()
```
